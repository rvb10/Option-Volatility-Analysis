```python
df = pd.read_excel('Option Chain.xlsx')
```

```python
df.head()
```

```python
pattern = r'^(?P<ticker>[A-Za-z]+)\s+(?P<expiry>\d{2}/\d{2}/\d{2})\s+(?P<option_type>[cCpP])(?P<strike>\d+\.?\d*)'
```

```python
parsed = df["ID_BB_SEC_NUM_DES"].str.extract(pattern)
```

```python
parsed["option_type"] = parsed["option_type"].str.upper()
parsed["strike"] = parsed["strike"].astype(float)
```

```python
parsed
```

```python
df = pd.concat([df, parsed], axis=1)
```

```python
df.head(10)
```

```python
df.to_excel("options_parsed.xlsx", index=False)
```
